transforming with sklearn

```python
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
```

```python
missing_df=pd.read_csv("datasets/car-sales-extended-missing-data.csv")
missing_df.dropna(subset=["Price"],inplace=True)
x=missing_df.drop('Price',axis=1)
y=missing_df['Price']
```

```python
x.isnull().sum()
```

```python
cat_imputer=SimpleImputer(strategy="constant",fill_value="missing")
door_imputer=SimpleImputer(strategy="constant",fill_value=4)
num_imputer=SimpleImputer(strategy="mean")
imputer =ColumnTransformer([("cat",cat_imputer,["Make","Colour"]),("num",num_imputer,["Odometer (KM)"]),("door",door_imputer,["Doors"])],remainder="passthrough")
```

```python
imputed_x=imputer.fit(x)
```

```python
imputed_x
```

```python
imputed_x=imputer.fit_transform(x)
```

```python
x=pd.DataFrame(imputed_x,columns=x.columns.get_level_values(0))
```

```python
x
```

```python
x.isnull().sum()
```

```python
one_hot=OneHotEncoder()
transformer=ColumnTransformer([("onehot",one_hot,["Make","Colour","Doors"])],remainder="passthrough")
```

```python
x.columns
```

```python
tranformed_x=transformer.fit_transform(x)
```

```python
tranformed_x
```

```python
x_train,x_test,y_train,y_test=train_test_split(tranformed_x,y,test_size=0.2)
```

```python
model=RandomForestRegressor(n_estimators=10000)
model.fit(x_train,y_train)
model.score(x_test,y_test)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_curve
```

```python
print(cross_val_score(model,tranformed_x,y))
```
