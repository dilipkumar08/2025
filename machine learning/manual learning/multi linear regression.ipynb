## Multi Linear Regression

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

So, in this note we gonna compare the predictions btw predefined Linear Regression model vs manually created for multi linear case

```python
class Linear_Regression():
    def __init__(self):
        self.weights=None
        
    def predict(self,input):
        biasInput=np.c_[np.ones((input.shape[0],1)),input]
        return biasInput@self.weights

    def train(self,x,y):
        xbias=np.c_[np.ones((x.shape[0],1)),x]
        self.weights=np.linalg.inv(xbias.T@xbias)@xbias.T@y

```

```python
squareFoot = [850, 1200, 900, 1600, 1400, 1350, 1700, 950, 1800, 750, 
      1400, 1600, 850, 1250, 1350, 1700, 950, 1850, 900, 750, 
      1350, 850, 1600, 1250, 900, 950, 1750, 1400, 1900, 750, 
      1600, 1400, 1350, 900, 1750, 950, 850, 1900, 1250, 750, 
      950, 900, 1350, 1400, 1750, 1900, 1600, 750, 1250, 850]

bedrooms = [2, 3, 2, 4, 3, 3, 4, 2, 5, 1, 
      3, 4, 2, 3, 3, 4, 2, 5, 2, 1, 
      3, 2, 4, 3, 2, 2, 4, 3, 5, 1, 
      4, 3, 3, 2, 4, 2, 2, 5, 3, 1, 
      2, 2, 3, 3, 4, 5, 4, 1, 3, 2]

distance = [8, 5, 10, 3, 4, 6, 2, 9, 1, 12, 
      4, 3, 8, 6, 6, 2, 9, 1, 10, 12, 
      6, 8, 3, 5, 10, 9, 2, 4, 1, 12, 
      3, 4, 6, 10, 2, 9, 8, 1, 5, 12, 
      9, 10, 6, 4, 2, 1, 3, 12, 5, 8]

age = [5, 10, 4, 20, 15, 12, 25, 6, 30, 2, 
      15, 20, 5, 12, 12, 25, 6, 35, 4, 2, 
      12, 5, 20, 10, 4, 6, 25, 15, 40, 2, 
      20, 15, 12, 4, 25, 6, 5, 40, 10, 2, 
      6, 4, 12, 15, 25, 40, 20, 2, 10, 5]

price = [45, 60, 50, 85, 70, 65, 90, 48, 110, 35, 
     72, 85, 47, 62, 68, 92, 50, 115, 55, 33, 
     68, 45, 88, 62, 49, 50, 95, 70, 120, 30, 
     86, 72, 66, 48, 98, 55, 46, 125, 62, 28, 
     52, 49, 68, 72, 98, 130, 85, 25, 60, 45] 
```

```python
x=pd.DataFrame({"sqft":squareFoot,"bedroom":bedrooms,"distance":distance,"age":age})
```

```python
y=pd.DataFrame({"price":price})
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7,shuffle=True)
```

```python
model1=Linear_Regression()
model1.train(x_train,y_train)
predictions=model1.predict(x_test)
```

```python
plt.plot(predictions,linestyle='--',color='black',marker='x')
plt.plot(y_test,linestyle='dotted',color='green',marker='o')
```

```python
model2=LinearRegression()
```

```python
model2.fit(x_train,y_train)
```

```python
y_preds=model2.predict(x_test)
```

```python
plt.plot(y_preds,linestyle='--',color='black',marker='x')
plt.plot(y_test,linestyle='dotted',color='green',marker='o')
```
